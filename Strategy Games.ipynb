```python
free_games = df[df["Free"]]
paid_games = df[df["Free"] == False]
plt.hist(free_games["Average User Rating"], alpha=0.5, label="Free games", color="blue", density=True)
plt.hist(paid_games["Average User Rating"], alpha=0.5, label="Paid games", color="red", density=True)
plt.title('Average User Rating of Free and Paid Games')
plt.xlabel('Average User Rating')
plt.ylabel('User Rating Count')
plt.legend()
plt.show()
```

As visualised in the graphic, free games have higher ratings than paid games between the 4.2-5 variables. So that we can say free games are higher rated by users.
